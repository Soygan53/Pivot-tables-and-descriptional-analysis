If we want to learn something about data, we need to look at data info firstly.

```python
"""
Created on Monday Nov 03 12:21:11 2022

@author: Ahmet Sina Soykan
"""

import pandas as pd                                 #importing libraries

data=pd.read_excel("Temporary_data3_-_Copy.xlsx")   #read the file
data.info()                                         #look at the info of the data
```

After that we need to look at null values.

```python
print(data.isnull().sum().sort_values(ascending=False)) #look at the nulls of the data
```

Than we describe the data.

```python
pd.options.display.max_rows = 999        # Display settings for descriptions
pd.options.display.max_columns = 999 
print(data.describe())                   # Describe the data
```
